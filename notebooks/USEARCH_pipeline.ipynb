* unzip files

```python
%%bash -s "$seqDir"
cd $1

for i in *.gz; do pigz -k -d -p 10 "$i" & done
    #find . maxdepth 1 -name '*.gz' -print0 #| xargs -0 -I {} -P 10 unzip {}
     #for i in *.gz; do echo "$i"; done
```

* move to working directory

```python
%%bash -s "$seqDir" "$workDir"
cd $2

mv $1/*fastq .
```

```python
%%bash -s "$workDir" "$Date"
cd $1
#mkdir merged
# Merge paired reads
# Add sample name to read label (-relabel option)
# Pool samples together in raw.fq (Linux cat command)

usearch -fastq_mergepairs *R1*.fastq -fastqout $2.merged.fq  -relabel @


```

```python
%%bash -s "$workDir" "$Date" 
cd $1
head -n 8 $2.merged.fq
```

```python
%%bash -s "$workDir" "$Date"
cd $1
usearch -fastx_truncate $2.merged.fq -stripleft 19 -stripright 20  \
  -fastqout $2.stripped.fq
```

## Quality filtering

```python
%%bash -s "$workDir" "$Date" "$maxee"
cd $1

usearch -fastq_filter $2.stripped.fq -fastq_maxee $3 \
  -fastaout $2.filtered.fa -relabel Filt
```

```python
%%bash -s "$workDir" "$Date" "$maxee" 
cd $1
head -n 8 $2.stripped.fq
```

```python
%%bash -s "$workDir" "$Date" "$maxee" 

cd $1

printf "Max expected error cutoff:"
echo $3

printf "Number of sequence pre-filter: "
grep -c "@" $2.stripped.fq

printf "Number of sequences post-filter: "
grep -c ">" $2.filtered.fa

echo '' 
#head -n 8 $1$2
head -n 8 $2.filtered.fa
```

# Remove sequences with "N"
apparently not necessary as no sequences filtered

```python
%%bash -s "$workDir" "$Date"

cd $1

bioawk -c fastx '{if ($seq !~ /N/){print ">" $name " " $4 "\n" $seq}}' $2.filtered.fa > $2.filtered.noN.fa

printf "Number of sequence pre-filter: "
grep -c ">" $2.filtered.fa

printf "Number of sequences post-filter: "
grep -c ">" $2.filtered.noN.fa
```

## Find unique sequences and abundances

```python
%%bash -s "$workDir" "$Date"
cd $1

usearch -fastx_uniques $2.filtered.fa -sizeout -relabel Uniq -fastaout $2.uniques.fa
```

```python
%%bash -s "$workDir" "$Date"

cd $1

printf "Number of unique sequences: "
grep -c ">" $2.uniques.fa

head $2.uniques.fa -n 10
```

# align to reference alignment

```python
%%bash -s "$workDir"
# filtering column positions in silva db

cd ~/databases

mothur "#filter.seqs(vertical=t, \
    fasta=silva.seed_v132.align, \
    processors=10)" > /dev/null

printf "Number of sequences post-filter: "
grep -c ">" silva.seed_v132.filter.fasta
```

```python
%%bash -s "$workDir" "$Date"
# aligning sequences

cd $1

mothur "#align.seqs(candidate=$2.uniques.fa, \
    template=~/databases/silva.seed_v132.filter.fasta, \
    processors=10, \
    flip=T)" | head -n 100
```

```python
%%bash -s "$workDir" "$Date"
# filtering out gap positions in the alignment

cd $1

mothur "#filter.seqs(vertical=t, \
    fasta=$2.uniques.align, \
    processors=10)" 
```

```python
%%bash -s "$workDir" "$Date"
# summarize alignment

cd $1

mothur "#summary.seqs(fasta=$2.uniques.filter.fasta, \
    processors=10)" 
```

## Removing homopolymers (> 8) and screening out sequences that don't align to an amplicon region

```python
%%bash -s "$workDir" "$Date"

cd $1

mothur "#screen.seqs(fasta=$2.uniques.filter.fasta, \
    processors=10, \
    start=566, \
    end=1480, \
    maxhomop=8, \
    minlength=50)" | tail -n 50  
```

```python
%%bash -s "$workDir" "$Date"

cd $1

printf "Number of sequences post-filter: "
grep -c ">" $2.uniques.filter.good.fasta

head $2.uniques.filter.good.fasta -n 20
```

## Cluster OTUs at 97%

```python
%%bash -s "$workDir" "$Date"
cd $1
usearch -cluster_otus $2.uniques.filter.good.fasta -otus $2.v2.otus.fa -relabel Otu
```

## Denoise: predict biological sequences and filter chimeras

```python
%%bash -s "$workDir" "$Date"
cd $1
usearch -unoise3 $2.uniques.fa -zotus $2.zotus.fa
```

## Make OTU table

```python
%%bash -s "$workDir" "$Date"
cd $1
usearch -otutab $2.merged.fq -otus $2.v2.otus.fa -otutabout otutab_raw.txt
```

```python
%%bash -s "$workDir" "$Date"
cd $1
head otutab_raw.txt
```

```python
%%bash -s "$workDir" "$Date"
cd $1
usearch -otutab $2.merged.fq -otus $2.zotus.fa -otutabout zotutab_raw.txt
```

# Assign taxonomy
* use the SILVA v123 LTP named isolate subset (12k seqs.) downloaded from https://www.drive5.com/usearch/manual/sintax_downloads.html on 20180130.

```python
%%bash -s "$workDir" "$Date" "$databaseDir"
cd $1
usearch -sintax $2.v2.otus.fa -db $3/rdp_16s_v16_sp.fa -strand both \
  -tabbedout sintax.txt -sintax_cutoff 0.8
```

```python
%%bash -s "$workDir" "$Date"
cd $1
usearch -sintax_summary sintax.txt -otutabin otutab_raw.txt -rank d -output phylum_summary.txt
```
