```r
library(dplyr)
library(ggplot2)
library(tidyr)
library(reshape2)
```

# Read blast table
* subset to just 99% matches

```r
Bdf = read.delim('/home/be68/Hyphosphere/data/3Exp/Isolate_ASV_blast.out', sep = "\t", header=FALSE) 

colnames(Bdf) = c('qseqid','sseqid',
'pident',
'length',
'mismatch',
'gapopen',
'qstart',
'qend',
'sstart',
'send',
'evalue',
'bitscore')

Bdf = Bdf %>%
    filter(., pident > 99, length > 250)
Bdf %>% head()
Bdf %>% dim()
```

## get list of ASVs that are close match to isolates

```r
Imatch = Bdf$sseqid %>% as.character()
```

# Read log2foldchange table

```r
LFC2 = read.delim('~/Hyphosphere/data/3Exp/DESeq/dfl2fc-Exp2.txt', sep = '\t') %>% filter(OTU %in% Imatch)
head(LFC2)
```

```r
LFC2B = left_join(LFC2, Bdf, by = c('OTU' ="sseqid"))
dim(LFC2B)
```

# Explore Burkholderia

```r
LFC2B %>%
    filter(Order == "Betaproteobacteriales" & Contrast == 'CH_CS') %>%
    select(OTU, Kingdom, Phylum, Class, Order, Family, Genus, qseqid, pident, length)%>%
    unique()
           
BurkToGraph =  
    LFC2B %>%
    filter(Order == "Betaproteobacteriales" & Contrast == 'CH_CS') %>%
    select(OTU, Kingdom, Phylum, Class, Order, Family, Genus, qseqid, pident, length)%>%
    unique() %>% 
    .$OTU %>% 
    as.character()
#            & Contrast %in% c('CH_CS', "BH_BS", "RH_RS")) %>% 
#     .$OTU %>% 
#     as.character %>% 
#     .
```

```r
LFC2B %>%
    filter(OTU %in% BurkToGraph & Contrast == 'CH_CS') %>%
    unique() %>%
    mutate(sig = ifelse(padj < 0.05, 'sig', 'ns')) %>%
    ggplot(aes(x = log2FoldChange, y = OTU)) +
        geom_point(aes(size = baseMean, shape = sig), fill = 'red') +
        scale_shape_manual(values = c(19, 21))+
        facet_wrap(~Subset)
```
