# Check which myxos are abundant in Exp1

```r
Test = subset_samples(ps, SampleType == "CH" & Soil != "Sand") %>%
        subset_taxa(., Order == "Myxococcales")
```

```r
TopMyxos = taxa_sums(Test) %>% sort(., decreasing = TRUE) %>% names() %>% .[1:20]
ps.snorm = transform_sample_counts(ps, function(x) x/sum(x))
Test.2 = subset_samples(ps.snorm, SampleType == "CH") %>%
            prune_taxa(TopMyxos, .)
Test.2
```

```r
tax_table(Test.2)
```

```r
otu_table(Test.2)
```

# Check which Fibrobacteraless are abundant in Exp1

```r
Test = subset_samples(ps, SampleType == "CH" & Soil != "Sand") %>%
        subset_taxa(., Order == "Fibrobacterales")
Test
```

```r
tax_table(Test.2)
```

```r
otu_table(Test.2)
```
